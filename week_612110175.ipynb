```python
userInput = input("Sample numbers :  ")
userList = userInput.split(",")
even = [item for item in userList if int(item)%2 == 0]
odd  = [item for item in userList if int(item)%2 != 0]
print("Number of even numbers : "+str(len(even)))
print("Number of odd numbers : "+str(len(odd)))
```
