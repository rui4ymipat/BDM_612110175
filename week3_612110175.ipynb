```python
# add data
titanics = np.genfromtxt('/kaggle/input/titanic.txt',delimiter='\t',missing_values = None,dtype=str)
titanics = np.delete(titanics,[1,6,7,8,9,10],1)
titanics = np.delete(titanics,2,1)
```

```python
#process
plt.subplot(1,2,1)  
plt.bar(np.unique(titanics[1:,1]),[sum(titanics[:,1]=='Female'),sum(titanics[:,1]=='Male')])
plt.subplot(1,2,2)
plt.bar(np.unique(titanics[1:,4]),[sum(titanics[:,4]=='No'),sum(titanics[:,4]=='Yes')])
```
